```python
import matplotlib.pyplot as plot
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans 
```

```python
df = pd.read_csv('Mall_Customers.csv')
clustering_data = df[["AnnualIncome","SpendingScore"]]
clustering_data
```

```python
x = clustering_data.values
x
```

```python
plot.scatter(clustering_data.AnnualIncome.to_list() , clustering_data.SpendingScore.to_list())
plot.title("Mall customers Spending Score")
plot.xlabel("Annual Income")
plot.ylabel("Spending Score")
plot.show()
```

```python
def get_wcss(X):
    wcss_list= []
    for i in range(1, 11):  
        kmeans = KMeans(n_clusters=i, init='k-means++', random_state= 42)  
        kmeans.fit(X)  
        wcss_list.append(kmeans.inertia_)  
    
    return wcss_list
```

```python
wcss = get_wcss(x)
[print([i+1,wcss[i]]) for i in range(10)]
plot.plot(range(1, 11), wcss)
plot.title('The Elbow Method Graph')  
plot.xlabel('Number of clusters(k)')  
plot.ylabel('wcss_list')
plot.show()    
```

```python
def clustering_kmeans(X,k):
    kmeans = KMeans(n_clusters=k, init='k-means++', random_state= 42)  
    y= kmeans.fit_predict(X)  
    return kmeans,y
```

```python
# Using the Elbow method the optimal value of cluster(k) is 5 for the given dataset
k_means, y = clustering_kmeans(x, 5)
```

```python
plot.scatter(x[y == 0, 0], x[y == 0, 1], s = 100, c = 'blue', label = 'Cluster 1') #for first cluster  
plot.scatter(x[y == 1, 0], x[y == 1, 1], s = 100, c = 'green', label = 'Cluster 2') #for second cluster  
plot.scatter(x[y== 2, 0], x[y == 2, 1], s = 100, c = 'red', label = 'Cluster 3') #for third cluster  
plot.scatter(x[y == 3, 0], x[y == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4') #for fourth cluster  
plot.scatter(x[y == 4, 0], x[y == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5') #for fifth cluster 
plot.scatter(k_means.cluster_centers_[:, 0], k_means.cluster_centers_[:, 1], s = 300, c = 'yellow', label = 'Centroid')   
plot.title('Customers Cluster')  
plot.xlabel('Annual Income')  
plot.ylabel('Spending Score')  
plot.legend()  
plot.show()  
```
